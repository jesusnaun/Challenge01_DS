```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url, engine="pyarrow")
tienda2 = pd.read_csv(url2, engine="pyarrow")
tienda3 = pd.read_csv(url3, engine="pyarrow")
tienda4 = pd.read_csv(url4, engine="pyarrow")

tienda.head()
```

#1. Análisis de facturación



```python
total1 = tienda["Precio"].sum()
total2 = tienda2["Precio"].sum()
total3 = tienda3["Precio"].sum()
total4 = tienda4["Precio"].sum()
```

```python
print(f"Facturación de la tienda 1: S/{total1:,.2f}")
print(f"Facturación de la tienda 2: S/{total2:,.2f}")
print(f"Facturación de la tienda 3: S/{total3:,.2f}")
print(f"Facturación de la tienda 4: S/{total4:,.2f}")
```

# 2. Ventas por categoría

```python
def dame_ventas_por_categoria(df, nombre_tienda):
    ventas=df.groupby("Categoría del Producto")["Precio"].sum()
    ventas = ventas.sort_index()

    print(f"Ventas por categoría en {nombre_tienda}: ")
    for categoria, total in ventas.items():
        print(f"{categoria}: S/{total:,.2f}")
    print()
```

```python
dame_ventas_por_categoria(tienda, "Tienda 1")
dame_ventas_por_categoria(tienda2, "Tienda 2")
dame_ventas_por_categoria(tienda3, "Tienda 3")
dame_ventas_por_categoria(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio1 = tienda["Calificación"].mean()
calificacion_promedio2 = tienda2["Calificación"].mean()
calificacion_promedio3 = tienda3["Calificación"].mean()
calificacion_promedio4 = tienda4["Calificación"].mean()
```

```python
print(f"Calificación promedio de la tienda 1: {calificacion_promedio1:,.2f}")
print(f"Calificación promedio de la tienda 2: {calificacion_promedio2:,.2f}")
print(f"Calificación promedio de la tienda 3: {calificacion_promedio3:,.2f}")
print(f"Calificación promedio de la tienda 4: {calificacion_promedio4:,.2f}")
```

# 4. Productos más y menos vendidos

```python
def top_productos(df, nombre_tienda):
    contador = df["Producto"].value_counts()

    top_mas = contador.head(3)
    top_menos = contador.tail(3)

    print(f"Ranking de {nombre_tienda}:")

    print("--Top 3 productos más vendidos--")
    for producto, cantidad in top_mas.items():
        print(f"{producto} : {cantidad} unidades")

    print("--Top 3 productos menos vendidos--")
    for producto, cantidad in top_menos.items():
        print(f"{producto} : {cantidad} unidades")

    print()
```

```python
top_productos(tienda, "Tienda 1")
top_productos(tienda2, "Tienda 2")
top_productos(tienda3, "Tienda 3")
top_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
costoEnvio_promedio1 = tienda["Costo de envío"].mean()
costoEnvio_promedio2 = tienda2["Costo de envío"].mean()
costoEnvio_promedio3 = tienda3["Costo de envío"].mean()
costoEnvio_promedio4 = tienda4["Costo de envío"].mean()
```

```python
print(f"Costo de envío promedio de la tienda 1: S/{costoEnvio_promedio1:,.2f}")
print(f"Costo de envío promedio de la tienda 2: S/{costoEnvio_promedio2:,.2f}")
print(f"Costo de envío promedio de la tienda 3: S/{costoEnvio_promedio3:,.2f}")
print(f"Costo de envío promedio de la tienda 4: S/{costoEnvio_promedio4:,.2f}")
```

# 6. Facturación - costo de envío por tienda

```python
totalreal1 = tienda["Precio"].sum() - tienda["Costo de envío"].sum()
totalreal2 = tienda2["Precio"].sum() - tienda2["Costo de envío"].sum()
totalreal3 = tienda3["Precio"].sum() - tienda3["Costo de envío"].sum()
totalreal4 = tienda4["Precio"].sum() - tienda4["Costo de envío"].sum()
```

```python
print(f"Facturación real de la tienda 1: S/{totalreal1:,.2f}")
print(f"Facturación real de la tienda 2: S/{totalreal2:,.2f}")
print(f"Facturación real de la tienda 3: S/{totalreal3:,.2f}")
print(f"Facturación real de la tienda 4: S/{totalreal4:,.2f}")
```

```python
import plotly.express as px
```

```python
labels = ["Tienda 1","Tienda 2", "Tienda 3", "Tienda 4"]
values = [totalreal1,totalreal2,totalreal3, totalreal4 ]

graficoPie = px.pie(names=labels, values=values, title="Participación de cada tienda")
graficoPie.show()
```

# 7. Ticket promedio por tienda

```python
promedioreal1 = totalreal1 / len(tienda)
promedioreal2 = totalreal2 / len(tienda2)
promedioreal3 = totalreal3 / len(tienda3)
promedioreal4 = totalreal4 / len(tienda4)
```

```python
print(f"Ticket promedio de la tienda 1: S/{promedioreal1:,.2f}")
print(f"Ticket promedio de la tienda 2: S/{promedioreal2:,.2f}")
print(f"Ticket promedio de la tienda 3: S/{promedioreal3:,.2f}")
print(f"Ticket promedio de la tienda 4: S/{promedioreal4:,.2f}")
```

```python
import plotly.express as px
```

```python
labels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
values = [promedioreal1, promedioreal2, promedioreal3, promedioreal4]

graficoBarras = px.bar(
    x=labels,
    y=values,
    text=values,
    title="Promedio de ticket por tienda",
    labels={"x": "Tienda", "y": "Promedio de ticket"},
    color=labels
)

graficoBarras.update_traces(texttemplate='$%{text:.2f}', textposition='outside')
graficoBarras.update_layout(
    yaxis_tickprefix="$",
    xaxis_title="Tiendas",
    yaxis_title="USD",
    yaxis_range=[350000,max(values)*1.1])

graficoBarras.show()
```

# 8. Ventas vs Cantidad de pedidos por tienda

```python
ventas = [totalreal1, totalreal2, totalreal3, totalreal4]
ventas_escaladas = [v/100000 for v in ventas]
pedidos = [len(tienda), len(tienda2), len(tienda3), len(tienda4)]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Bar(
    x=tiendas,
    y=ventas_escaladas,
    name="Monto de ventas",
    marker_color="green",
    text=[f"${v:,.2f}" for v in ventas_escaladas],
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=tiendas,
    y=pedidos,
    name="Cantidad de pedidos",
    marker_color="orange",
    text=pedidos,
    textposition='outside'
))

fig.update_layout(
    barmode='group',
    title="Monto de ventas vs Cantidad de pedidos por tienda",
    xaxis_title="Tiendas",
    yaxis_title="Valores",
    yaxis_tickprefix="",
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)

fig.show()
```

# 9. Análisis del desempeño geográfico

```python
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"


ventas_geo = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

#ventas_geo.rename(columns={"lat": "Latitud", "lon": "Longitud"}, inplace=True)

fig = px.scatter_map(
    ventas_geo,
    lat="lat",
    lon="lon",
    color="Tienda",
    size="Precio",  # opcional: para reflejar monto
    hover_data=["Producto", "Precio","Tienda"],
    zoom=3,
    height=600,
    title="Ventas geolocalizadas por tienda"
)

# Usar mapa base de Mapbox (requiere token o usar estilo open-source)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

fig.show()
```
